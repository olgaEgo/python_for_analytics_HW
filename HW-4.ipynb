### 1.1. Скачать данные по ссылке https://www.kaggle.com/datasets/ionaskel/laptop-prices

```python
df = pd.read_csv('./kc-house-data.csv', encoding='1251')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(20, 10))
sns.histplot(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость недвижимости')
plt.ylabel('Количество домов')
plt.ticklabel_format(style='plain')
df['price'].describe();
```

Вывод: Стоимость большей части домов не превышает 1 млн.

### 1.3 Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['sqft_living'])
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Количество домов');
```

Вывод: Жилая рлощадь большинства домов находится в пределах от 1000 до 3000 квадратных метров.

### 2.1.4 Изучите распределение года постройки

```python
df['yr_built'].max()
```

```python
plt.figure(figsize=(20, 10))
sns.histplot(df['yr_built'], bins=23)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.xticks(np.arange(1900, 2020, step=5));
```

Выводы: Сильный спад строительства пришелся на период с 1930 по 1940 гг, а так же с 1970 по 1975. Подъем строительства наблюдается в 1940г, в период с 1975 по 1980 гг и с 2000 по 2010 гг.

## Задача 2

### 2.1 Изучите распределение домов от наличия вида на набережную
- Постройте график
- Сделайте выводы

```python
df_waterfront = df['waterfront'].value_counts()
df_waterfront.index = ['без вида', 'с видом']
df_waterfront.head()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(df_waterfront, autopct='%1.1f%%')
plt.title('Наличие вида на набережную')
plt.legend(df_waterfront.index);
```

Вывод: График показывает, что только 0,8% домов имеют вид на набережную.

### 2.2 Изучите распределение этажей домов

```python
plt.figure(figsize=(10, 6))
plt.hist(df['floors'], align='left')
plt.title('Распределение домов по количеству этажей')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов');
```

Вывод: На графике видно, что большая часть домов имеют 1 или 2 этажа.

### 2.2 Изучите распределение состояния домов

```python
df_cond = df['condition'].value_counts().reset_index()
df_cond
```

```python
plt.figure(figsize=(6, 4))
# plt.hist(df['condition'])
sns.barplot(x=df_cond['index'], y=df_cond['condition'])
plt.title('Распределение домов по состоянию')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов');
```

## 3 задача
### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Вывод: Наблюдается прямая линейная зависимость между стоимостью и площадью домов.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Вывод: Наблюдается прямая линейная зависимость между стоимостью и уровенем конструкции и дизайна зданий.

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), showfliers=False)
plt.title('Влияние этажности на стоимость здания')
plt.xlabel('Стоимость')
plt.ylabel('Количество этажей');
```

Вывод: Наибольшую стоимость иимеют дома с количеством этажей равным 2,5.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(
    x=df['price'], y=df['condition'].astype('str'),
    order=('1', '2', '3', '4', '5'),
    showfliers=False)
plt.title('Влияние состояния на стоимость здания')
plt.xlabel('Стоимость')
plt.ylabel('Состояние');
```

Вывод: Оценка состояния домов 3 и выше увеличивает стоимость домов.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['yr_built'], y=df['price'])

plt.title('Зависимость стоимости здания от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость')
plt.xticks(np.arange(1900, 2020, step=10));
```

Вывод: Стоимость здания не зависит от года его постройки. Дома, построенные в период с 1945 по 1990, имеют в среднем меньшую стоимость по сравнению с остальными.
